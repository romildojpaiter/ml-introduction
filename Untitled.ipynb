```python
import numpy as np

# Inicialização das dimensões
n_entradas = 10
n_escondida = 50
n_saida = 3

# Inicialização dos pesos com He initialization
W_h = np.random.randn(n_entradas, n_escondida) * np.sqrt(2 / n_entradas)
W_saida = np.random.randn(n_escondida, n_saida) * np.sqrt(2 / n_escondida)

# Inicialização dos limiares de ativação (bias) como zeros
b_h = np.zeros(n_escondida)
b_saida = np.zeros(n_saida)

print("Matriz de Pesos da Camada Escondida (W_h):")
print(W_h)
print("Limiar de Ativação da Camada Escondida (b_h):")
print(b_h)
print("Matriz de Pesos da Camada de Saída (W_saida):")
print(W_saida)
print("Limiar de Ativação da Camada de Saída (b_saida):")
print(b_saida)
```
